```python
import graphlab
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
new_cust = graphlab.SFrame.read_csv('WH_model_1_061416_v2.csv')
depositors = new_cust[new_cust['deposit_flag']==1]
```

```python
print len(depositors)
depositors.show(view='Scatter Plot',x='BetAmount',y='gross_win')
```

```python
def statement_cat(SF):
    if SF == 'Never':
        return 'Never'
    else:
        return 'Weekly/Monthly/Fortnightly'
depositors['SF_cat'] = depositors['StatementFrequency'].apply(statement_cat)
depositors.show(view='Bar Chart',x='SF_cat',y='gross_win')
```

```python
betcount = depositors[depositors['BetCount']<150]
betcountshow = betcount[betcount['gross_win']<10000]
betcountshow.show(view='Scatter Plot',x='BetCount',y='gross_win')
```
